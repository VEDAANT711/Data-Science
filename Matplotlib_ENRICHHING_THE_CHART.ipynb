```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_val = np.arange(-10,10,.5)
y_val = x_val**3
plt.xticks(np.arange(-10,10,1),rotation=90)
plt.plot(x_val,y_val)
plt.show()
```

```python
x_val = np.arange(-4,4,.1)
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val)
plt.show()
```

```python
x_val = np.arange(-4,4,.1)
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val,color="yellow")
plt.show()
```

```python
x_val = np.arange(-4,4,.1) 
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val,color="yellow",marker = '*')
plt.show()
```

```python
x_val = np.arange(-4,4,.1) 
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val,'g*')
plt.show()
```

```python
x_val = np.arange(-4,4,.1) 
plt.xticks(np.arange(-4,4,.3),rotation=90)


y1_val = x_val**2
plt.plot(x_val,y1_val)

y2_val  = x_val**3
plt.plot(x_val,y2_val)


plt.show()
```

```python
x_val = np.arange(-4,4,.1)
plt.xticks(np.arange(-4,4,.3),rotation=90)


y1_val = x_val**2
y2_val = x_val**3

plt.plot(x_val,y1_val,label='square')
plt.plot(x_val,y1_val,label='cube')

plt.legend()
plt.show()
```

```python
x_val = np.arange(-4,4,.1) 
plt.xticks(np.arange(-4,4,.3),rotation=90)


y1_val = x_val**2
y2_val  = x_val**3


plt.plot(x_val,y1_val,label="Square")
plt.plot(x_val,y2_val,label="Cube")

plt.legend()
plt.show()
```

```python
x_val = np.arange(-4,4,.1) 
plt.xticks(np.arange(-4,4,.3),rotation=90)


y1_val = x_val**2
y2_val  = x_val**3


plt.plot(x_val,y1_val,)
plt.plot(x_val,y2_val,)

plt.legend(['Square','Cube'],loc='lower right',frameon=False) 
plt.show()
```

```python
x_val = np.arange(-4,4,.1) 
plt.xticks(np.arange(-4,4,.3),rotation=90)


y1_val = x_val**2
y2_val  = x_val**3


plt.plot(x_val,y1_val,)
plt.plot(x_val,y2_val,)

plt.legend(['Square','Cube'],loc='lower right',frameon=False,ncol=2) 
plt.show()
```

```python
x_val = np.arange(-4,4,.1) #elementwise operation possible
plt.xticks(np.arange(-4,4,.3),rotation=90)


y1_val = x_val**2
y2_val  = x_val**3


plt.plot(x_val,y1_val,)
plt.plot(x_val,y2_val,)

plt.legend(['Square','Cube'],loc='lower right',frameon=False,ncol=2) #chnaging the default to lower right

plt.text(0,0+.5,'(0,0)')
plt.text(1,1+.5,'two')


plt.xlim(-2,2)
plt.ylim(-2,2)

plt.show()
```
