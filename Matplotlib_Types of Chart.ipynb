```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.bar([1,2,3],[10,20,5])
plt.show()
```

```python
plt.bar([1,2,3],[10,20,5],width = 0.5)
plt.legend('shop1')
plt.show()
```

```python
plt.bar([1,2,3],[10,20,5]) 
plt.bar([1,2,3],[5,10,6],width=0.5)
plt.legend(["Shop1",'Shop2'])
plt.show()
```

```python
plt.bar([1,2,3],[10,20,5],width=0.3) 
plt.bar([1.3,2.3,3.3],[5,10,6],width=0.3)
plt.legend(["Shop1",'Shop2'])
plt.show()
```
